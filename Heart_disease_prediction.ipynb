## Model Evaluation

```python
x=df.iloc[:,0:-1] #independent variable
y=df.AHD #target variable
```

```python
#balance x and y data
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x_smote,y_smote=sm.fit_resample(x,y)
```

```python
x_smote.shape #shape
```

```python
y_smote.shape #shape
```

```python
#split train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_smote,y_smote,test_size=0.25,random_state=42)
```

```python
#apply LogisticRegression
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression() 
LR.fit(x_train,y_train)#fit train data 
```

```python
print(LR.intercept_) #line intercept
```

```python
y_pred=LR.predict(x_test) #y_pred
```

```python
y_train_predict=LR.predict(x_train) #y_train_predict
```

```python
#performance metrices
#accuracy score for train data
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_train,y_train_predict)
acc
```

```python
#performance metrices
#accuracy score for test data
from sklearn.metrics import accuracy_score,f1_score
acc=accuracy_score(y_test,y_pred)
acc
```

```python
#classification report
#test data
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#classification report
#train data
from sklearn.metrics import classification_report
print(classification_report(y_train,y_train_predict))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
#dataframe for confusion matrix
pd.crosstab(y_test,y_pred)
```

```python
y_prob=LR.predict_proba(x_test)[:,1] # predict proba for class 1
y_prob
```

```python
from sklearn.metrics import roc_auc_score,roc_curve,plot_roc_curve 
fpr,tpr,threshold=roc_curve(y_test,y_prob) 
```

```python
def plot_roc_curve(fpr, tpr):# function to plot roc curve
    plt.plot(fpr, tpr, color='orange', label='ROC')#line plot between fpr and tpr
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')# assigning name to  x axis
    plt.ylabel('True Positive Rate')# assigning name to y axis
    plt.title('Receiver Operating Characteristic (ROC) Curve')#assigning name to curve  
    plt.legend()#area describing the elements of the graph
    plt.show()#to show graph without location
```

```python
plot_roc_curve(fpr, tpr)#plotting ruc curve
```
