# model selection

```python
x=df[['Age','Experience','Income','ZIP_Code','Family','CCAvg','Education','Mortgage','Personal_loan','Security_Account','CD_Account','Online','CreditCard']]
x
```

```python
y=df.Personal_loan
y
```

```python
#split train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
#balance data
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_smote,y_smote=smote.fit_resample(x_train,y_train)
```

```python
#svc
from sklearn.svm import SVC
sc=SVC()
model=sc.fit(x_smote,y_smote)
```

```python
model
```

```python
#ytest_predict
y_test_pred=sc.predict(x_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
matrix_confusion = confusion_matrix(y_test,y_test_pred)
sns.heatmap(matrix_confusion, square=True, annot=True, cmap='Blues', fmt='d', cbar=False)
```

```python
#crosstab
cm=pd.crosstab(y_test,y_test_pred)
cm
```

```python
#accuracy_score
from sklearn.metrics import accuracy_score,classification_report,f1_score
acc=accuracy_score(y_test,y_test_pred)
acc
```

```python
#f1_core
f1=f1_score(y_test,y_test_pred)
f1
```

```python
#classification_report
print(classification_report(y_test,y_test_pred))
```

```python
#cross_val_score
from sklearn.model_selection import cross_val_score
scores = cross_val_score(sc,x_smote,y_smote,cv=3,scoring='f1')
print(scores)
print("Cross validation Score:",scores.mean())
print("Std :",scores.std())
#std of < 0.05 is good. 
```

# Hyperparameter

```python
#parameter 
params={'C':[0.1,1,10,100,1000],
        'gamma':[0.1,0.01,0.001,0.0001]}
```

```python
#fit in gridsearchcv
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(SVC(random_state=42),params,refit=True,scoring='f1',verbose=3,cv=3)#cv =3
grid.fit(x_smote,y_smote)
```

```python
print(grid.best_estimator_)#best_estimator
```

```python
print(grid.best_params_)#best_params
```

```python
clf=SVC(C=1000,gamma=0.0001,random_state=42)#best_params to svc
```

```python
clf.fit(x_smote,y_smote)
```

```python
y_pred_clf=clf.predict(x_test)#predict test data
```

```python
from sklearn.metrics import classification_report,accuracy_score
acc=accuracy_score(y_test,y_pred_clf)#accuracy_score
acc
```

```python
f1=f1_score(y_test,y_pred_clf)#f1_score
f1
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
matrix_confusion = confusion_matrix(y_test,y_pred_clf)
sns.heatmap(matrix_confusion, square=True, annot=True, cmap='Blues', fmt='d', cbar=False)
```

```python
pd.crosstab(y_test,y_pred_clf)
```

```python
print(classification_report(y_test,y_pred_clf))
```

```python
#cross validation after best_param
from sklearn.model_selection import cross_val_score
cross_validation=cross_val_score(clf,x_smote,y_smote,scoring='f1',cv=3)
print("Cross validation Score:",cross_validation.mean())
print("Std :",cross_validation.std())
```
